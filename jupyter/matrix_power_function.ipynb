```python
import numpy as np
from numpy.linalg import matrix_power

matrix = np.array([[0.8, 0.2],[0.3,0.7]])

print(f"n=2: {matrix_power(matrix, 2)}")
print(f"n=4: {matrix_power(matrix, 4)}")
print(f"n=8: {matrix_power(matrix, 8)}")
print(f"n=16: {matrix_power(matrix, 16)}")
print(f"n=100: {matrix_power(matrix, 100)}")
```
